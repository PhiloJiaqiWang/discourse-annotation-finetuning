```python
import pandas as pd

df = pd.read_csv("Qwen result.csv")  
df = df.dropna(how="all")

df
```

```python
from sklearn.metrics import cohen_kappa_score

k_qwen = cohen_kappa_score(df["gold_label"], df["qwen_label"])
k_lora = cohen_kappa_score(df["gold_label"], df["lora_label"])

print("Kappa (Human vs Qwen zeroshot):", round(k_qwen, 3))
print("Kappa (Human vs LoRA-Qwen):   ", round(k_lora, 3))
print("Δ Kappa:", round(k_lora - k_qwen, 3))
```

```python
from sklearn.metrics import confusion_matrix

def confusion_01(gold, pred):
    # labels=[0,1] 
    tn, fp, fn, tp = confusion_matrix(
        gold, pred, labels=[0, 1]
    ).ravel()
    return {
        "TP": tp,  # 1 → 1
        "FP": fp,  # 0 → 1
        "FN": fn,  # 1 → 0
        "TN": tn,  # 0 → 0
    }

conf_qwen = confusion_01(df["gold_label"], df["qwen_label"])
conf_lora = confusion_01(df["gold_label"], df["lora_label"])

print("Qwen zeroshot:", conf_qwen)
print("LoRA Qwen:   ", conf_lora)
```
